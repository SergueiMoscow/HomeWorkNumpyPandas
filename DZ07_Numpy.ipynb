```python
def conv2d(img: np.ndarray, f: np.ndarray) -> np.ndarray:
    res = []
    for i in range(3):
        a = img[:, :, i]
        s = f.shape + tuple(np.subtract(a.shape, f.shape) + 1)
        subM = np.lib.stride_tricks.as_strided(a, shape = s, strides = a.strides * 2)
        res.append(np.einsum('ij,ijkl->kl', f, subM))
    res = np.stack(res, axis=-1)
    return res
    
```

```python
conv_filter = np.full((9, 9), 1./81)
new_img = conv2d(img, conv_filter)
```
